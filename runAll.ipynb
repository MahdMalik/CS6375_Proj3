```python
from mlp import MLP_Manager
from cnn import CNN_Manager

complexities = ["low", "medium", "high"]

datasetTypes = ["mnist", "cifar"]

mlpModel = MLP_Manager()

cnnModel = CNN_Manager()

for dataset in datasetTypes:
    for complexityLevel in complexities:
        mlpModel.createFullModel(complexityLevel,dataset, "MLP")

for dataset in datasetTypes:
    for complexityLevel in complexities:
        cnnModel.createFullModel(complexityLevel,dataset, "CNN")
```
